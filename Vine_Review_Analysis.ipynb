```python
import pandas as pd

file = 'vine_table.csv'
vine = pd.read_csv(file)
vine
```

```python
vine = vine.loc[vine['total_votes'] >= 20]
```

```python
vine
```

```python
vine_filtered = vine.loc[vine['helpful_votes'] / vine['total_votes'] >= 0.5]
vine_filtered
```

```python
paid_reviews = vine_filtered.loc[vine_filtered['vine'] == 'Y']
paid_reviews
```

```python
unpaid_reviews = vine_filtered.loc[vine_filtered['vine'] == 'N']
unpaid_reviews
```

```python
paid_reviews_count = paid_reviews['review_id'].count()
paid_reviews_count
```

```python
unpaid_reviews_count = unpaid_reviews['review_id'].count()
unpaid_reviews_count
```

```python
paid_5star_reviews = paid_reviews.loc[paid_reviews['star_rating'] == 5]
paid_5star_reviews_count = paid_5star_reviews['star_rating'].count()
paid_5star_reviews_count
```

```python
unpaid_5star_reviews = unpaid_reviews.loc[unpaid_reviews['star_rating'] == 5]
unpaid_5star_reviews_count = unpaid_5star_reviews['star_rating'].count()
unpaid_5star_reviews_count
```

```python
paid_5star_per = str(round(((paid_5star_reviews_count / paid_reviews_count)*100),1)) + '%'
paid_5star_per
```

```python
unpaid_5star_per = str(round(((unpaid_5star_reviews_count / unpaid_reviews_count)*100),1)) + '%'
unpaid_5star_per
```

```python
index = ['Total Reviews', '5 Star Reviews', '5 Star %']
review_analysis = pd.DataFrame(columns=['Paid Reviews', 'Unpaid Reviews'], index=index)
```

```python
review_analysis['Paid Reviews'] = [paid_reviews_count, paid_5star_reviews_count, paid_5star_per]
review_analysis['Unpaid Reviews'] = [unpaid_reviews_count, unpaid_5star_reviews_count, unpaid_5star_per]
```
